```python
elements = []

while True:
    
    print("Elemens in our list:")
    for element in elements:
        print(" -> " + element)
        
    print("\nOptions:")
    print("1. Add Element.")
    print("2. Delete Element")
    print("3. Delete All Elements")
    print("4. Exist")
    print("Note: Please give input in numbers e.g 1, 2, 3, 4")
    
    inp = input("Enter Your choice: ")
    
    if inp == '1':
        elements.append(input("Enter Your Element: "))
    elif inp == '2':
        elements.remove(input("Enter Element To Delete"))
    elif inp == '3':
        elements.clear()
    elif inp == '4':
        break
    else:
        print("Invalid Input")
    print("\nxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx\n")

```
